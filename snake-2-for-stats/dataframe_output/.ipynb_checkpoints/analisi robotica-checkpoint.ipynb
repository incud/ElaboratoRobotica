```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; 
sns.set()
sns.set(font_scale=2)
```

```python
opt_inter = pd.read_csv("df_latest.csv")
opt_end = pd.read_csv("df_latest_nomiddlepoints.csv")
tracik = pd.read_csv("df_ros_2.csv")
opt_timing = pd.read_csv("df_latest_nomiddlepoints_allmethods.csv") # only 32 joints
```

```python
t1 = opt_timing[["method", "time"]]
t2 = tracik.query("model == 'modello_32.urdf'")[["method", "time"]]
df_timing = t1.append(t2, ignore_index=True, sort=False)
```

```python
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(11.7, 8.27))
sns.set_color_codes("pastel")
sns.barplot(x="method", y="time", data=df_timing, ci=None, orient="v")
ax.set_title("Confronto tempi medi di calcolo della cinematica inversa")
ax.set(xlabel='Algoritmo utilizzato', ylabel='Tempo (s)')
```

```python
df_timing_1 = opt_timing.groupby(["method"]).time.describe().reset_index()
df_timing_2 = tracik.query("model == 'modello_32.urdf'").groupby(["method"]).time.describe().reset_index()
df_timing = df_timing_1.append(df_timing_2, ignore_index=True, sort=False)
df_timing
```

```python
df_timing_tr = df_timing.drop(columns=["method", "count"]).transpose().rename(columns={
    0: "COBYLA",
    1: "L-BFGS-B",
    2: "SLSQP",
    3: "TNC", 
    4: "TRAC_IK"
}).reset_index().drop(columns=["index"])
df_timing_tr
```

```python
fig, ax = plt.subplots(figsize=(11.7, 8.27))
sns.set_color_codes("pastel")

data = df_timing_tr
```

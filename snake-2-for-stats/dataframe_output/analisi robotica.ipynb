```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; 
sns.set()
sns.set(font_scale=2)
```

```python
opt_inter = pd.read_csv("df_latest.csv")
opt_end = pd.read_csv("df_latest_nomiddlepoints.csv")
tracik = pd.read_csv("df_ros_2.csv")
opt_timing = pd.read_csv("df_latest_nomiddlepoints_allmethods.csv") # only 32 joints
```

```python
tracik.query("model == 'modello_8.urdf'").query("main_point_inferred == '[0 0 0]'")
```

```python
import math 
x = 0.142758 
y = 0.074665 
z = 0.009396
math.sqrt(x**2 + y**2 + z**2)
```

```python
opt_end.query("model == 'modello_8.urdf'").query("method == 'SLSQP'")
```

```python
tracik = tracik.replace(to_replace={
    "modello_8.urdf": 8,
    "modello_16.urdf": 16,
    "modello_24.urdf": 24,
    "modello_32.urdf": 32,
    "modello_40.urdf": 40,
    "modello_48.urdf": 48,
})
opt_end = opt_end.replace(to_replace={
    "modello_8.urdf": 8,
    "modello_16.urdf": 16,
    "modello_24.urdf": 24,
    "modello_32.urdf": 32,
    "modello_40.urdf": 40,
    "modello_48.urdf": 48,
})
```

```python
t1 = opt_timing[["method", "time"]]
t2 = tracik.query("model == 32")[["method", "time"]]
df_timing = t1.append(t2, ignore_index=True, sort=False)
```

```python
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(11.7, 8.27))
sns.set_color_codes("muted")
sns.barplot(x="method", y="time", data=df_timing, ci=None, orient="v")
ax.set_title("Confronto tempi medi di calcolo della cinematica inversa", fontweight="bold", pad=20)
ax.set(xlabel='Algoritmo utilizzato', ylabel='Tempo (s)')
```

```python
fig, ax = plt.subplots(figsize=(11.7, 8.27))
sns.set_color_codes("deep")
ax = sns.lineplot(x="model", y="main_point_error (%)", data=opt_end.query("method == 'SLSQP'"), ci="sd", legend='brief', label="SLSQP")
ax = sns.lineplot(x="model", y="main_point_error (%)", data=opt_end.query("method == 'COBYLA'"), ci="sd", legend='brief', label="COBYLA")
ax = sns.lineplot(x="model", y="main_point_error (%)", data=tracik.query("method == 'trac-ik'"), ci="sd", legend='brief', label="TRAC-IK")
ax.set(ylim=(0, 100))
ax.set(xticks=range(8, 49, 8))
ax.set_title("Errore percentuale sul calcolo della cinematica inversa (media, std)", fontweight="bold", pad=20)
ax.set(xlabel='Numero di giunti', ylabel='Errore (%)')
```

```python
t1 = opt_end[["model", "method", "main_point_error (%)"]]
t2 = tracik[["model", "method", "main_point_error (%)"]]
df_correctness = t1.append(t2, ignore_index=True, sort=False).rename(columns={"main_point_error (%)":"error"})
df_correctness["error_desc"] = ""
df_correctness.loc[df_correctness["error"] < 0.1,           "error_desc"] = "< 0.1"
df_correctness.loc[df_correctness["error"].between(0.1,1),  "error_desc"] = "0.1 - 1"
df_correctness.loc[df_correctness["error"].between(1,10),   "error_desc"] = "1 - 10"
df_correctness.loc[df_correctness["error"].between(10,100), "error_desc"] = "> 10"
df_correctness
```

```python
fig, ax = plt.subplots(figsize=(11.7, 8.27))
ax = sns.countplot(x="error_desc", hue="method", data=df_correctness.query("model == 8"), order=["< 0.1", "0.1 - 1", "1 - 10", "> 10"])
ax.set_title("Accuratezza dei punti (8 giunti)", fontweight="bold", pad=20)
ax.set(xlabel='Errore percentuale', ylabel='Numero di punti')
```

```python
fig, ax = plt.subplots(figsize=(11.7, 8.27))
ax = sns.countplot(x="error_desc", hue="method", data=df_correctness.query("model == 24"), order=["< 0.1", "0.1 - 1", "1 - 10", "> 10"])
ax.set_title("Accuratezza dei punti (24 giunti)", fontweight="bold", pad=20)
ax.set(xlabel='Errore percentuale', ylabel='Numero di punti')
```

```python
fig, ax = plt.subplots(figsize=(11.7, 8.27))
ax = sns.countplot(x="error_desc", hue="method", data=df_correctness.query("model == 48"), order=["< 0.1", "0.1 - 1", "1 - 10", "> 10"])
ax.set_title("Accuratezza dei punti (48 giunti)", fontweight="bold", pad=20)
ax.set(xlabel='Errore percentuale', ylabel='Numero di punti')
```
